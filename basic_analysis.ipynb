```python
import pandas as pd
pd.options.display.max_rows = 999

# Read the data
data = pd.read_csv("data/change2017_full-extract_clean_github.csv")
data = data.rename(columns={'Quel âge avez-vous ?': 'Âge', 'Quelle est votre tranche de revenu annuel ?': 'Revenu',
                        'Quelle est la taille de votre commune ?': 'Commune',
                        'À quelle tendance politique vous identifiez-vous?': 'Tendance',
                        'Quelle est votre profession ?':'Profession', 'Vous êtes :':'Genre',
                        'Dans quel département habitez vous ?': 'Departement'})
```

```python
# Select some variables
variable_list = ['Âge', 'Revenu', 'Commune', 'Tendance', 'Profession', 'Genre', 'Departement']

# For every variables, count the value, then reindex, then print
for variable_name in variable_list:
    variable = data[variable_name].value_counts()
    if variable_name == 'Âge':
        variable = variable.reindex(["16- 20", "21-25","26-30", "31-35","36-40",  
         "41-45", "46-50", "51-55", "56-60", "61 -65","66-70" ,"71-75", "76-80",
         "Plus de 80 ans","Je préfère ne pas répondre"])
    elif variable_name == 'Revenu':
        variable = variable.reindex(["Jusqu'à 9 700 euros","de 9 700 à 26 791 euros",
                                     "de 26 791 à 71 826 euros","de 71 826 à 152 108 euros",
                                     "au-delà de 152 108 euros","Je préfère ne pas répondre"])                         
    elif variable_name == 'Commune':
        variable = variable.reindex(["Moins de 2000 habitants","2000 à 20 000 habitants",
                                     "20 000 à 100 000 habitants","Plus de 100 000 habitants",
                                     "Je préfère ne pas répondre"])
    elif variable_name == 'Tendance':
        variable = variable.reindex(["Extrême droite","Droite","Centre-droite",
                                     "Centre",  "Centre-gauche","Gauche", "Extrême gauche",  "Écologiste",
                                     "Autre tendance", "Je préfère ne pas répondre"])
    elif variable_name ==  'Genre':
        variable = variable.reindex(["Un homme", "Une femme", "Autre", "Je préfère ne pas répondre"])
    elif variable_name == "Profession":
        variable = variable.reindex(["Cadres et professions intellectuelles supérieures",
                                      "Employés", "Professions Intermédiaires", "Etudiant", "Sans emploi",
                                      "Artisans, commerçants et chefs d’entreprise", "Ouvriers", "Lycéen",
                                      "Agriculteurs exploitants", "Autre", "Je préfère ne pas répondre"])
    elif variable_name == "Departement":
        variable = variable.sort_index()
    print(variable_name)
    print(len(variable_name) * "¯" + "\n")
    print(variable)
    print("\n" + '#' * 40 + "\n")
```
